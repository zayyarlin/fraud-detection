```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import lightgbm as lgb
import seaborn as sns
sns.set(style="whitegrid")
```

## Data pre-processing

```python
train_identity = pd.read_csv('data/train_identity.csv')
train_transaction = pd.read_csv('data/train_transaction.csv')

# Combine 
train = pd.merge(train_transaction, train_identity, on='TransactionID', how='left')
train.to_csv('data/train.csv')
```

```python
train = pd.read_csv('data/train.csv')
```

```python
len(train)
```

```python
train.shape
```

```python
target_count = train.isFraud.value_counts()
```

```python
target_count.plot(kind="bar", title="Fraud ratio")
```

Since the data set is very unbalanced, we will resample

```python
# Class count
count_not_fraud, count_fraud = train.isFraud.value_counts()

# Divide by class
df_class_not_fraud = train[train['isFraud'] == 0]
df_class_fraud = train[train['isFraud'] == 1]
```

```python
df_class_not_fraud_under = df_class_not_fraud.sample(count_fraud)
df_under = pd.concat([df_class_not_fraud_under, df_class_fraud], axis=0)

print('Random under-sampling:')
print(df_under.isFraud.value_counts())

df_under.isFraud.value_counts().plot(kind='bar', title='Count (is Fraud)');
```

```python
# set the categorical values
categorical_columns = ['ProductCD','addr1', 'addr2','P_emaildomain','R_emaildomain']
categorical_columns += ['card' + str(i) for i in range(1,7)] # card1-card6
categorical_columns += ['M' + str(i) for i in range(1,10)] # M1-M9
categorical_columns += ['DeviceType','DeviceInfo']
categorical_columns += ['id_' + str(i) for i in range(12,39)] # id12 - id38


for col in categorical_columns:
    df_under[col] = df_under[col].astype('category')
```

## Visualization

```python
df_under_numeric = df_under[df_under.columns.difference(categorical_columns)]
df_under_numeric = df_under_numeric.fillna(0)
```

```python
from sklearn.decomposition import PCA
```

```python
# Create a smaller data set for viz
df_viz = df_under_numeric.sample(frac=0.1)
```

```python
X_pca = df_viz[df_viz.columns.difference(['isFraud'])]
```

```python
y_pca = df_viz['isFraud']
```

```python
pca = PCA(n_components=2)
pca.fit(X_pca)
X = pca.transform(X_pca)
```

```python
plt.figure(figsize=(16, 6))
sns.scatterplot(x = X[:, 0],y= X[:, 1], hue = y_pca)
```

```python
from sklearn.manifold import TSNE
X_embedded = TSNE(n_components=2, perplexity=10).fit_transform(X_pca)
```

```python
plt.figure(figsize=(16, 6))
sns.scatterplot(x = X_embedded[:, 0],y= X_embedded[:, 1], hue = y_pca)
```

## Test-Train Split

```python
from sklearn.model_selection import train_test_split

train_under, test_under = train_test_split(df_under, test_size = 0.2)

X_train = train_under[train_under.columns.difference(['isFraud'])]
y_train = train_under['isFraud']

X_test = test_under[test_under.columns.difference(['isFraud'])]
y_test = test_under['isFraud']
```

```python
del train_identity, train_transaction
```

```python
del df_under, train, df_class_not_fraud, df_class_fraud, train_under, test_under
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## Classification using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Using only numerical columns
X_train_lr = X_train[X_train.columns.difference(categorical_columns)]
X_train_lr = X_train_lr.fillna(0)
```

```python
X_train_lr.shape
```

```python
clf = LogisticRegression(random_state=0, solver='lbfgs', multi_class='ovr', max_iter=1000).fit(X_train_lr, y_train)
```

```python
# Using only numerical columns
X_test_lr = X_test[X_test.columns.difference(categorical_columns)]
X_test_lr = X_test_lr.fillna(0)
```

```python
X_test_lr.shape
```

```python
clf.score(X_test_lr, y_test)
```

The logistics regression accuracy is pretty low: 70%

## Classification using Gradient Boosted Decision Tree

```python
import lightgbm as lgb
lgb_train = lgb.Dataset(X_train, y_train)
```

```python
params = {'num_leaves': 256,
          'min_child_samples': 79,
          'objective': 'binary',
          'max_depth': 13,
          'learning_rate': 0.03,
          "boosting_type": "gbdt",
          "subsample_freq": 3,
          "subsample": 0.9,
          "bagging_seed": 11,
          "metric": 'binary_error',
          "verbosity": -1,
          'reg_alpha': 0.3,
          'reg_lambda': 0.3,
          'colsample_bytree': 0.9
         }

print('Starting training...')
# train
#gbm = lgb.cv(params, lgb_train, num_boost_round=100, nfold=5)
gbm = lgb.train(params,
                lgb_train,
                num_boost_round=100)
```

```python
y_pred = gbm.predict(X_test, num_iteration=gbm.best_iteration)
```

```python
y_pred_binarized = [1 if x >= 0.5 else 0 for x in y_pred]
```

```python
accuracy = (y_test == y_pred_binarized).mean()
```

```python
accuracy
```

The Gradient boosted tree accuracy is around 86% on test data set. This could be made better with hyper-parameter tuning.
